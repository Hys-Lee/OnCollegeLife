```python
data  ## 데이터에 강남과 강남구가 있음 ->이런걸 하나로 만들어야 : Cleaning해야함
         # Cleaning이 데이터 분석의 대부분을 차지할 것임 대부분 데이터는 지저분하기 때문임
    ## 영등포구 여의도동 이므로 사실 가장 BEST는 강남을 강남구로 바꾸자.
    # 데이터에 가이드라인을 만들어서 하나의 원칙을 따르도록 하는게 좋다
```

```python
data['지점'].replace("강남","강남구").replace("여의도","여의도동") # replace

# pandas의 대부분의 기능은 원본을 바꾸지 않음
```

```python
data['지점'].replace("강","강남구")#replace기능은 일부 글자를 바꾸는게 아니라 특정 데이터 자체를 바꾸는 것임 -> 안바뀜

# 그러나 문자열 기능에서 replace는 문자열 일부를 바꿀 수 있는 기능이 있음

# 따라서 이 기능을 쓰기 위해
data['지점'].str.replace("강","강남구") #이렇게 문자열 기능을 str.을 통해 사용가능
```

```python
## unique:값의 종류, nunique:값의 종류 개수, value_counts:값들의 각각 개수
data['지점'].unique()
data['지점'].nunique()
data['지점'].value_counts()
data['지점'].value_counts(normalize=True) #비율로
```

```python
# NaN=Not a Number=Null=NA=Blank != 'NaN' 그니까 문자열은 아니다
data['몸무게'].fillna(0) #NaN을 0으로 채우기
data['몸무게'].dropna() #NaN을 없애기
```

## 조건부 검색

```python
data[data['몸무게']>76]  ## True False 만든 리스트를 검색하면 됨->True가 검색되지
data.loc[data['몸무게']>76] # 원래 True인 행을 검색하는거니 loc를 써야 하긴 하는데, 조건부 검색은 많이 사용해서 이를 유일하게 생략가능

## 꿀팁 : 아무거나 쓴거를 드래그 하고 괄호나 따옴표누르면 사이에 들어감(근데 여는거로 해야함)
```

```python
data[data['지점']=='강남']
```

```python
# 1. 담당이 김지수가 아닌 데이터 검색
data[data['담당']!='김지수']
```

```python
# 2. 측정일이 2019-01-04 이후인 데이터 검색
data[data['측정일']>"2019-01-04"] # 문자열의 사전식 배열 생각
```

```python
data[(data['몸무게']>76) & (data['담당']=='김지수')]    # 파이썬에서도 &사용함, and는 bool에서 쓰는건가봐

kim=data['담당']=='김지수'
over_weight=data['몸무게']>76

cols=['이름','측정일']
data.loc[kim&over_weight,cols] ## 이렇게 분리하는게 더 좋은 습관이래
```

## 행열 추가/수정!!!

```python
##추가나 수정은 코드가 같음

data['날씨']='맑음'     ## 있으면 덮어 씌운다는 느낌, 없으면 만든다는 느낌
data                     ## 원본이 바뀐다 =가 있으면 그런 느낌임
data['날씨']='흐림'
data
```

```python
#행열 추가할 때 길이만 행이나 열 길이만 같으면 들어감
data['과체중']=data['몸무게']>76
cols=['과체중','몸무게']
data[cols] # NaN이면 False가 되기는 한다
```

```python
# 문자열 넣자
# 과체중(76초과)과 정상체중
# 각 조건의 데이터를 각각 검색하고 상태란 칼럼을 추가한다

over_weight=data['몸무게']>76

data.loc[over_weight,'상태']='과체중' ## 행열 동시 검색을 이용해 덮어쓴다고 생각(만든다고)

# ~:NOT을 이용해 그 반대의 상태 만들기는
data.loc[~over_weight,'상태']='정상체중'
cols=['몸무게','과체중','상태']
data[cols]
```
